```python
# TOURS DE HANOÏ

def hanoi(n, origin, inter, destination):
    if n==1:
        print(origin + "->" + destination)
    else:
        hanoi(n-1, origin, destination, inter)
        print(origin + "->" + destination)
        hanoi(n-1, inter, origin, destination)

hanoi(7, "A", "B", "C")
```

```python
# CALCUL SOMME D'UN TABLEAU

def sum(t, s=0, i=0):
    if(len(t)-1 == i):
        return s+t[i]
    else:
        return sum(t, s+t[i], i+1)

print(sum([42,73,1337]))
```

```python
# Triangle de Sierpinsky


import sys, pygame
pygame.init()

w,h = 1024, 768
screen = pygame.display.set_mode((w,h))
print(pygame.display.get_window_size())


color = (0, 0, 0)
# ICI ON DEFINIT DES TRUCS

def mid(m, n):
    mx,my = m
    nx, ny = n

    return (mx+(nx-mx)/2, my+(ny-my)/2)

def triangle(n, a, b, c):
    if(n==1):
        pygame.draw.aaline(screen, color, a, b)
        pygame.draw.aaline(screen, color, b, c)
        pygame.draw.aaline(screen, color, c, a)
    else:
        ax,ay = a
        bx,by = b
        cx,cy = c
        triangle(n-1, a, mid(a,c), mid(a,b))
        triangle(n-1, b, mid(b,a), mid(b,c))
        triangle(n-1, c, mid(a,c), mid(b,c))

i = 1
my_font = pygame.font.SysFont('JetBrains Mono Regular', 80)
play = True
clock = pygame.time.Clock()

while play:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            play = False
        if event.type == pygame.MOUSEMOTION:
            pass
            # print(event.pos)
        if event.type == pygame.KEYUP:
            print(event.key, event.unicode, event.scancode)
            if event.key == pygame.K_ESCAPE:
                play = False

    screen.fill((255, 255,255))
    triangle(i, (200,400), (700,100), (600, 650))
    text_surface = my_font.render(str(i), False, (0, 0, 0))
    screen.blit(text_surface, (0,0))    
    if(i < 15):
        i+=1

    # ICI ON AFFICHE ET ON BOUGE DES TRUCS

    clock.tick(1)
    pygame.display.flip()
```

```python
# Arbre Binaire

class Tree:
    pass

class EmptyTree(Tree):
    def __init__(self):
        pass

    def countNode(self):
        return 0

    def indentedString(self, level=0):
        return ""

    def countLeaf(self):
        return 0

    def sumAllButLeaf(self):
        return 0

    def __str__(self):
        return ""

    def isEmpty(self):
        return True

    def total(self):
        return 0

class Tree:
    def __init__(self, value, left=EmptyTree(), right=EmptyTree()):
        self.value = value
        self.left = left
        self.right = right

    def countNode(self):
        total = 0
        total += 1
        total += self.left.countNode()
        total += self.right.countNode()

        return total

    def countLeaf(self):
        if self.left.isEmpty() and self.right.isEmpty():
            return 1
        else:
            return 0 + self.left.countLeaf() + self.right.countLeaf()

    def sumAllButLeaf(self):
        total = 0
        stack = [self]

        while stack:
            current = stack.pop()
            total += current.value
            if not current.left.isEmpty():
                stack.append(current.left)
            if not current.right.isEmpty():
                stack.append(current.right)

        return total

    def total(self):
        total = 0
        total += (self.value)
        total += (self.left.total())
        total += (self.right.total())

        return total

    def isEmpty(self):
        return False

    def __str__ (self):
        s = ""
        s += str(self.value) + " "
        s += self.left.__str__()
        s += self.right.__str__()

        return s

    def appendOrdonne(self, value):
        if(self.value >= value):
            if(self.left.isEmpty()):
                self.left = Tree(value)
            else:
                self.left.appendOrdonne(value)
        else:
            if(self.right.isEmpty()):
                self.right = Tree(value)
            else:
                self.right.appendOrdonne(value)

    def indentedString(self, level=0):
        s = ""
        s+= '-'*level + str(self.value) + '\n'
        s+= self.left.indentedString(level+1)
        s+= self.right.indentedString(level+1)

        return s


treeSimple = Tree(4, Tree(2), Tree(5))
treeBig = Tree(4, Tree(2, Tree(5, Tree(4))), Tree(5, Tree(8), Tree(6)))

# print("Arbre Simple " + str(treeSimple))
# print("Arbre Simple Total Noeud " + str(treeSimple.total()))
# print("Arbre Simple Nombre Noeud " + str(treeSimple.countNode()))
# print("Arbre Simple Nombre Feuille " + str(treeSimple.countLeaf()))
# print("Arbre Simple Somme Noeud != Feuille " + str(treeSimple.sumAllButLeaf()))
print("Arbre Simple Indenté \n" + str(treeSimple.indentedString()))
treeSimple.appendOrdonne(9)
print("Arbre Simple Indenté \n" + str(treeSimple.indentedString()))
print("------------------------------------------------------------")
# print("Arbre Complexe " + str(treeBig))
# print("Arbre Complexe Total Noeud " + str(treeBig.total()))
# print("Arbre Complexe Nombre Noeud " + str(treeBig.countNode()))
# print("Arbre Complexe Nombre Feuille " + str(treeBig.countLeaf()))
# print("Arbre Complexe Somme Noeud != Feuille " + str(treeBig.sumAllButLeaf()))
print("Arbre Simple Indenté \n" + str(treeBig.indentedString()))
treeBig.appendOrdonne(0)
print("Arbre Simple Indenté \n" + str(treeBig.indentedString()))
```
