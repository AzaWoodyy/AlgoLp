```python
### Méthode de chiffrement RSA

## Génération des clés

import random
class Keys:
    # Variables
    numberOfPrimeNb = 0
    primeNumbers = []
    ide = 0
    e = 0
    d = 0
    n = 0

    # Function to calculate and initialize all variables required to get the keys
    def __init__(self, nbPrime = 1000):
        self.numberOfPrimeNb = nbPrime
        self.primeNumbers = self.getNPrimeNumbers(2)
        self.idE = self.getIndicatriceEuler()
        self.e = self.findPrimeWith()
        self.d = self.modInverse()
        self.n = self.encryption_module()

    # Function to get 2 random prime numbers between a define range
    def getNPrimeNumbers(self, n):
        primeNumbers = self.getPrimes()
        returnedPrimeN = []
        for i in range(n):
            returnedPrimeN.append(primeNumbers[random.randint(0, len(primeNumbers)-1)])

        return returnedPrimeN

    # Function to get the a list of N prime numbers
    def getPrimes(self):
        primes = []
        i = 2
        while i < self.numberOfPrimeNb:
            isPrime = True
            for j in range(2,int(i**.5)+1):
                if i%j == 0:
                    isPrime = False
                    break
            if isPrime:
                primes.append(i)
            i+=1

        return primes

    # Function to get idE (indicatrice of Euler) ((primeN1 -1)(primeN2 -1))
    def getIndicatriceEuler(self):
        return (self.primeNumbers[0] - 1) * (self.primeNumbers[1] - 1)

    # Function to get the lowest coPrime number of a given number (idE)
    def findPrimeWith(self):
        e = 2
        while e <= self.idE:
            if self.is_coprime(e, self.idE):
                return e
            else:
                e += 1
        return self.idE

    # Function to check if 2 numbers are coprime
    def is_coprime(self, e, idE):
        return self.gcd(e, idE) == 1

    # Function to get the gcd of 2 numbers
    def gcd(self, e, idE):
        while idE != 0:
            e, idE = idE, e%idE
        return e

    # Function to get the modular multiplicative inverse of 2 coprime numbers (e * x === 1 % idE) 
    def modInverse(self):
        for d in range(1, self.idE):
            if((self.e % self.idE) * (d % self.idE) % self.idE==1):
                return d
        return 0

    # Function to get the encryption_module (primeN1 * primeN2)
    def encryption_module(self):
        return self.primeNumbers[0] * self.primeNumbers[1]

    # Function to get the public key (e, n)
    def get_public_key(self):
        return self.e, self.n

    # Function to get the private key (d, n)
    def get_private_key(self):
        return (self.d, self.n)

    # Function for displaying all variables and test
    def debug(self):
        print("-------------------------------------------------------------------------------")
        print("                        Detail values for debug                                ")
        print("-------------------------------------------------------------------------------")
        print("      Deux nombres premiers = " + str(self.primeNumbers))
        print("      Indicatrice d'Euler (idE) soit ("+str(self.primeNumbers[0])+"-1)("+str(self.primeNumbers[1])+"-1) = " + str(self.idE))
        print("      Nombre premier avec Indicatrice d'Euler e = " + str(self.e))
        print("      Is Indicatrice d'Euler and e CoPrime = " + str(self.is_coprime(self.e, self.idE)))
        print("      Inverse Multiplicative Modulaire de l'Indicatrice d'Euler et e = " + str(self.d))
        print("      Module d'encryption n = " + str(self.n))
        print("      Clé publique = " + str(self.get_public_key()))
        print("      Clé privée = " + str(self.get_private_key()))
        print("-------------------------------------------------------------------------------")


# Generate keys
rsa_keys = Keys()

# Print out detail of the algorithm
rsa_keys.debug()
```
